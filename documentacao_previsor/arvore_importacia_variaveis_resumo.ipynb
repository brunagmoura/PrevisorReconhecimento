#Análise de variáveis mais relevantes por meio de árvores de decisão


O principal objetivo da análise foi indicar quais as variáveis mais relevantes para cada conjunto de dados (e não propriamente obter um modelo otimizado). Assim, não foram testados diferentes hiperparâmteros.

Para cada modelo de dados (agregados/não agregados) e intervalo (2011 a 2017/ 2018 a 2024) aplicou-se:
- Modelo de árvore de decisão a todos os dados (apenas para fins exploratórios);
- Modelo de árvore de decisão (com treino e teste);
- Modelo de regressão logística (com remoção de todos os valores faltantes - o que pode ter levado a um baixo desempenho);
- Random forest.



##Observações gerais:

- É válido manter região e UF como variáveis?
- Prejuízos privados e públicos, bem como danos humanos e COBRADEs, aparecem com frequência.
- Para as árvores de decisão, o f1-score não variam muito entre os modelos com e sem dados agregados.
- Para a regressão logística, o f1-score é ligeiramente melhor nos modelos sem os dados agregados.
